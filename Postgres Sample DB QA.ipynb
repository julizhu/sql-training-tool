```python
%%sql
-- Querying rentals
-- Case of converting a date to a string and then aggregating to the month level
SELECT left(cast(rental_date as varchar), 7) as month, count(*)
from rental 
GROUP BY left(cast(rental_date as varchar), 7)
ORDER BY MONTH desc
```
